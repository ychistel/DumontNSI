```python
# Prénom masculins VS prénoms féminins
# Ce qui les différencie c'est le sexe correspondant à la première valeur de chaque enregistrement
# 1 pour garçon et 2 pour fille
import csv

def enregistrements_sans_doublons(table):
    t=[]
    prenom=''
    for ligne in table:
        if ligne[1]==prenom:
            continue
        else:
            t.append(ligne)
            prenom=ligne[1]
    return t
    
with open('nat2018.csv','r') as f:
    # on passe la première ligne correspondant aux descripteurs
    f.readline()
    # table est une variable qui contient chaque enregistrement sous forme de tableau
    table = csv.reader(f,delimiter=";")
    enr=enregistrements_sans_doublons(table)
    sexe='1'
    compteurM=0
    compteurF=0
    # on parcourt la variable enr avec une boucle for
    for r in enr:
        if r[0]=='1':
            compteurM=compteurM+1
        elif r[0]=='2':
            compteurF=compteurF+1
        else:
            continue
    print("Le fichier contient",compteurM,"prénoms masculins et",compteurF,"prénoms féminins")
    if compteurM>compteurF:
        print("Il y a plus de prénoms masculins")
    else:
        print("Il y a plus de prénoms féminins")
```

```python
# Écrire un algorithme qui rassemble dans un même dictionnaire tous les enregistrements d'un même prénom. 
#
# Dans un même dictionnaire :
import csv

def enregistrements_meme_prenom_dico(table,prenom,sexe):
    # on crée un dictionnaire vide
    t={}
    # on récupère les descripteurs de la table (ligne 1 : ['sexe','prénom','année','nombre'])
    descripteurs=table.fieldnames
    for ligne in table:
        if ligne[descripteurs[1]]==prenom.upper() and ligne[descripteurs[0]]==sexe:
            t[ligne[descripteurs[2]]]=[ligne[descripteurs[0]],ligne[descripteurs[1]],ligne[descripteurs[3]]]
        else:
            continue
    return t

# fonction pour avoir un affichage lisible
def affiche_enregistrements_dico(d):
    for clef in d.keys():
        print(clef,":",d[clef])

with open('nat2018.csv','r') as f:
    table = csv.DictReader(f,delimiter=";")
    # On initialise les données de la recherche dans 2 variables
    prenom="félix"
    sexe='1'
    # On affecte à la variable t le tableau contenant les enregistrements selon les critères prénom et sexe
    d=enregistrements_meme_prenom_dico(table,prenom,sexe)
    #print("Il y a",len(t),"enregistrements pour",prenom)
    affiche_enregistrements_dico(d)
```

```python
# 7. Écrire un programme qui renvoie l'année où le prénom a été le plus attribué.
import csv
       
with open('nat2018.csv','r') as f:
    table = csv.reader(f,delimiter=";")
    # On initialise les données de la recherche dans 2 variables
    prenom="félix"
    sexe='1'
    # On affecte à la variable t le tableau contenant les enregistrements selon les critères prénom et sexe
    t=enregistrements_meme_prenom(table,prenom,sexe)
    # recherche de la valeur maximale : on parcourt t et on mémorise la valeur max
    if t!=[]:
        M=eval(t[0][3])
        for k in range(1,len(t)):
            if M<eval(t[k][3]):
                M=eval(t[k][3])
                annee=t[k][2]
    #affiche_enregistrements(t)
    print("Au maximum dans une même année, le prénom",prenom,"a été attribué",M,"fois en",annee)
```

```python
# 8. Écrire un programme de calcul de la moyenne des nombres d'usage d'un même prénom.
import csv
       
with open('nat2018.csv','r') as f:
    table = csv.reader(f,delimiter=";")
    # On initialise les données de la recherche dans 2 variables
    prenom="félix"
    sexe='1'
    # On affecte à la variable t le tableau contenant les enregistrements selon les critères prénom et sexe
    t=enregistrements_meme_prenom(table,prenom,sexe)
    # calcul de la moyenne : on parcourt t et on calcule la somme des valeurs que l'on divise par 
    # le nombre total des valeurs ajoutées non nulles
    if t!=[]:
        s,n=0,0
        for k in range(0,len(t)):
            if eval(t[k][3])!=0:
                s=s+eval(t[k][3])
                n=n+1
    moyenne=int(s/n)
    print("Le prénom",prenom,"a été attribué, en moyenne",moyenne,"fois chaque année sur",n,"ans.")
    #affiche_enregistrements(t)
```

```python
# 9. Écrire un programme qui trie les données (sur un même prénom) dans l'ordre croissant du nombre d'usage.

import csv

def tri_par_insertion(t):
    j=1
    while j<len(t):
        i=j-1
        k=t[j]
        while i>=0 and eval(t[i][3])>eval(k[3]):
            t[i+1]=t[i]
            i=i-1
        t[i+1]=k
        j=j+1
    return t

with open('nat2018.csv','r') as f:
    table = csv.reader(f,delimiter=";")
    # On initialise les données de la recherche dans 2 variables
    prenom="félix"
    sexe='1'
    # On affecte à la variable t le tableau contenant les enregistrements selon les critères prénom et sexe
    t=enregistrements_meme_prenom(table,prenom,sexe)
    tri_par_insertion(t)
    affiche_enregistrements(t)
```
